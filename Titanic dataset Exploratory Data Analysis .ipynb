```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
sns.set_theme(style="whitegrid")
```

```python
df = pd.read_csv('C:\\Users\\user\\Downloads\\train.csv')
```

```python
df.head()
```

```python
# Shape of dataset
print(f"Dataset contains {df.shape[0]} rows and {df.shape[1]} columns.")
```

```python
# Dataset information
print("\n--- Info ---")
df.info()
```

```python
# Statistical summary
print("\n--- Describe ---")
print(df.describe())
```

```python
# Value counts for each column (only for categorical and few discrete features)
print("\n--- Value Counts for Categorical Columns ---")
for col in df.columns:
    if df[col].dtype == 'object' or df[col].nunique() < 20:
        print(f"\nValue counts for {col}:\n")
        print(df[col].value_counts())
```

```python
# ---  Missing Values ---

# Check missing values
print("\n--- Missing Values ---")
print(df.isnull().sum())
```

```python
# ---  Visual Exploration ---

# Histograms for numerical features
df.hist(figsize=(15, 12), bins=30)
plt.suptitle('Histograms of Numerical Features', fontsize=16)
plt.show()
```

```python
# Boxplots for each numerical feature
for col in df.select_dtypes(include=np.number).columns:
    plt.figure(figsize=(8, 4))
    sns.boxplot(x=df[col])
    plt.title(f'Boxplot of {col}')
    plt.show()
```

```python

# Scatterplots: Pairwise relationships
sns.pairplot(df.select_dtypes(include=np.number))
plt.suptitle('Pairplot of Numerical Features', y=1.02)
plt.show()
```

```python
# Heatmap of correlation matrix (numerical features only)
plt.figure(figsize=(12, 8))
corr = df.select_dtypes(include=['number']).corr()
sns.heatmap(corr, annot=True, fmt='.2f', cmap='coolwarm', square=True)
plt.title('Correlation Heatmap')
plt.show()
```

```python
# ---  Categorical Feature Exploration ---

# Bar plots for categorical features
categorical_cols = df.select_dtypes(include='object').columns

for col in categorical_cols:
    plt.figure(figsize=(8, 4))
    sns.countplot(y=col, data=df, palette='viridis')
    plt.title(f'Countplot of {col}')
    plt.show()
```

```python
# ---  Observations Section ---

print("\n--- Observations ---")
print("""
1. From .info() and missing value check, we can identify if we need data cleaning.
2. Histograms show the distribution of numerical features (e.g., normal distribution, skewness).
3. Boxplots help detect outliers in numeric features.
4. Pairplots reveal possible linear/non-linear relationships between numeric features.
5. The heatmap shows which features are highly correlated (positive or negative).
6. Bar plots show the frequency of categories in categorical columns.
""")
```
